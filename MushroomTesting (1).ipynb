We will be using SVC (Support Vector Classifier) from the Scikit-learn library to implement this model and evaluating the predictions using a classification report and a confusion matrix. We will later optimize the model using grid search.

```python
##import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV

##convert non-numeric variables
df1 = pd.read_csv('Mushroom_clean.csv')



from sklearn.preprocessing import LabelEncoder

# Define a list of features you want to encode
features_to_encode = ['class', 'cap-shape','cap-surface','cap-color','does-bruise-or-bleed','gill-attachment','gill-color','stem-color','has-ring','ring-type','habitat','season']

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode each feature in the DataFrame
for feature in features_to_encode:
    df1[feature] = label_encoder.fit_transform(df1[feature])
df1
# If you want to transform new data using the same encoder,
##Set x and y variables
df2=df1.head(10000)
df2.info()
```

```python
x=df2.drop('class', axis=1)
y=df2['class']

##Split the data into training and test tests
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.3, random_state=12)

##set algorithm
model=SVC()

##fit the algorith to the training data
model.fit(x_train, y_train)

#assign a new variable to the prediction of the x test data
model_predict=model.predict(x_test)

#confusion matrix
print(confusion_matrix(y_test, model_predict))

#classification report
print(classification_report(y_test, model_predict))


```

The confusion matrix reports a low occurance of false-positives (3), and the classification report states that precision, recall, and the f1-score are all above 0.97
