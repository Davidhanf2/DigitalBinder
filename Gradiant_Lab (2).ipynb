We will be using SVC (Support Vector Classifier) from the Scikit-learn library to implement this model and evaluating the predictions using a classification report and a confusion matrix. We will later optimize the model using grid search.

```python
##import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
df = pd.read_csv('BetterLVTripAdvisor.csv') #read the csv
df
```

```python



from sklearn.preprocessing import LabelEncoder

# Define a list of features you want to encode
features_to_encode = ['Hotel stars','Period of stay', 'Traveler type','Pool','Gym','Tennis court','Spa','Casino','Free internet','Hotel name','User continent','User country','Review month','Review weekday']

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode each feature in the DataFrame
for feature in features_to_encode:
    df[feature] = (label_encoder.fit_transform(df[feature]))
df
# If you want to transform new data using the same encoder,
##Set x and y variables
```

```python
x=df.drop('Score', axis=1)
y=df['Score']

##Split the data into training and test tests
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.4, random_state=15)

##set algorithm
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=.1, max_depth=3, random_state=15)
gbm.fit(x_train, y_train)
##fit the algorith to the training data

#assign a new variable to the prediction of the x test data
model_predict=gbm.predict(x_test)

#confusion matrix
print(confusion_matrix(y_test, model_predict))

#classification report
print(classification_report(y_test, model_predict))


```
